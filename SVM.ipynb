```python
import pickle
import pandas as pd
from sklearn.model_selection import train_test_split
from preprocess1 import *
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("loan_approved.csv")
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df.loc[df.Gender.isnull(),'Gender']="Male"
df.loc[df.Married.isnull(),'Married']="Yes"
df.loc[df.Dependents.isnull(),'Dependents']="0"
df.loc[df.Self_Employed.isnull(),'Self_Employed']="No"
df.loc[df.LoanAmount.isnull(),'LoanAmount']=df.LoanAmount.median()
df.loc[df.Loan_Amount_Term.isnull(),'Loan_Amount_Term']=df.Loan_Amount_Term.median()
df.loc[df.Credit_History.isnull(),'Credit_History']=0.0
```

```python
df.isnull().sum()
```

```python
with open("preprocessing1 (1).pkl","rb") as f:
    preprocessor=pickle.load(f)
```

```python
preprocessor
```

```python
train,test,_,_=train_test_split(df,df["LoanAmount"],test_size=0.2,random_state=42)
```

```python
train['Loan_Status (Approved)'].value_counts()
```

```python
processed_data=preprocessor.fit_transform(train)
```

```python
processed_data
```

```python
x_train=processed_data[:,:-1]
y_train=processed_data[:,-1]
y_train
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE()
```

```python
x_train_smote,y_train_smote=smote.fit_resample(x_train,y_train)
```

```python
len(x_train_smote)
```

```python
len(y_train_smote)
```

```python
len(y_train)
```

```python
from sklearn.svm import SVC
svclassifier = SVC()
svclassifier.fit(x_train_smote, y_train_smote)
```

```python
test_processed=preprocessor.transform(test)
```

```python
x_test=test_processed[:,:-1]
y_test=test_processed[:,-1]
y_test
```

```python
y_pred=svclassifier.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test,y_pred))
```

```python
#hyperparameters
from sklearn.model_selection import GridSearchCV
from itertools import product

param_grid_linear = {
    'C': [0.1, 5, 10, 15, 60, 70],
    'kernel' : ['linear'],
    'gamma' : ['scale', 'auto']
}
param_grid_rbf = {
    'C': [0.1, 5, 10, 15, 60, 70],
    'kernel' : ['rbf'],
    'gamma' : ['scale', 'auto']
}
param_grid_poly = {
    'C': [0.1, 5, 10, 15, 60, 70],
    'kernel' : ['poly'],
    'gamma' : ['scale', 'auto']
}
```

```python
model_linear = SVC()
model_rbf = SVC()
model_poly = SVC()
model = SVC()

grid = GridSearchCV(model, param_grid=param_grid_rbf,refit=True,verbose=2,scoring='f1',cv=3)
grid.fit(x_train_smote, y_train_smote)
```

```python
with open("model.pkl","wb") as f:
    pickle.dump(grid,f)
```

```python
print(grid.best_params_)
print(grid.best_score_)
```
